```python
#import dependencies
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


#Loading_dataset
df = sns.load_dataset("iris")

#printing 1st 5
print(df.head())
print(df.tail())

#Display Basic information
print(df.info())

#Basic stats

import numpy as np
print(df.describe())

#Missing value as Null
df.isnull().sum()

#Duplicate row
print(df.duplicated().sum())

#Categorical value
print(df['species'].value_counts())

#Visualize categorical count
plt.figure(figsize=(8,5))
sns.countplot(x='species',data=df)
plt.title ("Iris")
plt.show()

#Pair plot
sns.pairplot(df,hue='species')
plt.show()

#Correlation Matrix
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

#INDEX
df.set_index('species',inplace=True)
print(df.head())
```

# 4. the mean, median, and standard deviation of the 'sepal_length' column

```python
import seaborn as sns
import pandas as pd

# Load dataset
iris = sns.load_dataset("iris")

# Calculate mean, median, and standard deviation
mean_sepal_length = iris['sepal_length'].mean()
median_sepal_length = iris['sepal_length'].median()
std_sepal_length = iris['sepal_length'].std()

mean_sepal_length, median_sepal_length, std_sepal_length
```

# 5. Which species has the longest average 'petal_length'

```python
import seaborn as sns
import pandas as pd

# Load dataset
iris = sns.load_dataset("iris")

# Group by species and calculate average petal_length
avg_petal_length = iris.groupby('species')['petal_length'].mean()

# Display the result
avg_petal_length
```

# 6. Create a histogram of 'sepal_width'. What does the distribution look like (e.g., normal, skewed

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")

# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(iris['sepal_width'], kde=True, bins=20, color='blue')

plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# 7. Use a boxplot to compare the 'petal_length' across different species. What do you observe

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")

# Create boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='petal_length', data=iris, palette='pastel')

plt.title('Petal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.show()
```

# 8. Plot a pairplot of all the numerical columns. Do you observe any clear separation between species

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")

# Create pairplot
sns.pairplot(iris, hue='species', corner=True, palette='Set2')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```
