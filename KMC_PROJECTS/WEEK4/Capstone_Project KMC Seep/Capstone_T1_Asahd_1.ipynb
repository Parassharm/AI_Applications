```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#सुरुमा original र cleaned dataset को आकार तुलना गर्ने
original_shape = df.shape

#Duplicate हटाउने
df_cleaned = df.drop_duplicates()

#नयाँ आकार
cleaned_shape = df_cleaned.shape

#के कति हटियो भनेर देखाउने
print("Duplicate हटाउनु अघि:", original_shape)
print("Duplicate हटाएपछि:", cleaned_shape)
print("हटाइएका पंक्तिहरू:", original_shape[0] - cleaned_shape[0])
```

```python
#मर्नेको जानकारी निकाल्ने (9999-99-99 भनेको मरेका होइनन्)
df['IS_DEAD'] = df['DATE_DIED'].apply(lambda x: 0 if x == '9999-99-99' else 1)
```

```python
#अज्ञात मानहरूलाई NaN मा परिवर्तन गर्ने
cols_to_clean = ['ICU', 'PNEUMONIA', 'DIABETES', 'COPD', 'ASTHMA', 'INMSUPR',
                 'CARDIOVASCULAR', 'OBESITY', 'RENAL_CHRONIC', 'TOBACCO']
df[cols_to_clean] = df[cols_to_clean].replace([97, 98, 99], np.nan)
```

```python
#Missing values drop गर्ने (सजिलोको लागि)
df = df.dropna(subset=cols_to_clean)
```

```python
# Q1 (25th percentile) अनि Q3 (75th percentile) निकाल्दा
Q1 = df['AGE'].quantile(0.25)
Q3 = df['AGE'].quantile(0.75)
IQR = Q3 - Q1

# Outliers नभएका row हरू मात्र राख्दा
filtered_df = df[(df['AGE'] >= Q1 - 1.5 * IQR) & (df['AGE'] <= Q3 + 1.5 * IQR)]

# Outliers हटाइएको पछि boxplot परेर चेक गरौं
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(x='IS_DEAD', y='AGE', data=filtered_df)
plt.title("Age distribution after removing outliers")
plt.xlabel("Death/Survi")
plt.ylabel("उमेर")
plt.show()
```

```python
# Feature र Target छुट्याउने

# 'AGE', 'SEX', 'DIABETES', 'OBESITY', 'PNEUMONIA', 'TOBACCO' लाई इनपुट फिचरको रूपमा लिने
X = df[['AGE', 'SEX', 'DIABETES', 'OBESITY', 'PNEUMONIA', 'TOBACCO']]

# 'IS_DEAD' लाई लक्षित Variable (target) को रूपमा छुट्याउने
y = df['IS_DEAD']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# डाटा विभाजन
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# मोडेल बनाउने र तालिम दिने
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Accuracy हेर्ने
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# मोडेलले टेस्ट डाटामा के अनुमान गर्‍यो?
y_pred = model.predict(X_test)

# Confusion Matrix देखाउने (कुन-कुन prediction सही/गलत भयो)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification Report (precision, recall, f1-score सहित विस्तृत रिपोर्ट)
print("Classification Report:\n", classification_report(y_test, y_pred))

# Accuracy Score देखाउने (कुल correct predictions को प्रतिशत)
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

'Positive' भन्यो (जस्तै – मरेको), तिनीहरूमध्ये कति साँचोमा 'Positive' थिए
वास्तवमा 'Positive' (मरेका) भएका मध्ये कति जनालाई मोडेलले सही रूपमा पत्ता लगायो

```python
#आवश्यक लाइब्रेरी इंपोर्ट
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Iris dataset लोड गर्ने
iris = load_iris()
X = iris.data[:, :2]  # केवल sepal length अनि sepal width लियौं
y = iris.target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.20,
                                                    random_state=42)

# Logistic Regression Model लगाउने
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

# Plotting decision boundary
# meshgrid
x_vals, y_vals = np.meshgrid(np.linspace(X[:,0].min()-1, X[:,0].max()+1, 500),
                             np.linspace(X[:,1].min()-1, X[:,1].max()+1, 500))

# हर एक point मा predict गरौं
Z = model.predict(np.c_[x_vals.ravel(), y_vals.ravel()])
Z = Z.reshape(x_vals.shape)

# ग्राफ Plot गरौं
plt.contourf(x_vals, y_vals, Z, cmap='coolwarm', alpha=0.8)
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap='coolwarm', edgecolor='black')
plt.scatter(X_test[:,0], X_test[:,1], c=y_test, cmap='coolwarm', marker='x')
plt.xlabel("Sepal Lenght")
plt.ylabel("Sepal Width")
plt.title("Iris Logistic Regression Decision Boundary")
plt.show()
```
